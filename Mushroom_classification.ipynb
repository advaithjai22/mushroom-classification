Importing the libraries 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

Reading the file and examining the data

```python
data = pd.read_csv("C:\\Users\\jadva\\OneDrive\\Desktop\\Datasets\\mushroom\\mushrooms.csv")
data.head()
```

Checking the data for null values

```python
print(data.isnull().sum())
```

We see that 'class' can take two unique values,'p' and 'e' denoting poisonous and edible respectively

```python
data['class'].unique()
```

Determining the shape of the dataframe

```python
data.shape
```

Using LabelEncoder to encode the categorical variables

```python
data.select_dtypes(exclude=['int', 'float']).columns
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
data.head()
```

```python
print(data.groupby('class').size())
```

Splitting the dataset into dependent and independent variables

```python
X = data.iloc[:,1:23]  
y = data.iloc[:, 0]  
X.head()
y.head()
```

```python
X.describe()
```

Checking for correlation between the various features

```python
data.corr()
```

Plotting a correlation matrix to visualize the correlation between the different features

```python
corr=data.corr()
sns.set(rc={'figure.figsize':(20,20)})
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           annot=True)
```

standaradizing the numerical values in the DataFrame

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X=scaler.fit_transform(X)
X
```

Splitting the dataset into test and train sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Using default Logistic Regression to fit the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn import metrics

model_LR= LogisticRegression()
```

```python
model_LR.fit(X_train,y_train)
```

Predicting the outputs using Logistic Regression

```python
y_pred = model_LR.predict(X_test)
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
model_LR.score(X_test, y_pred)
```

Calculating Area under the curve for the classifier

```python
auc_roc=metrics.roc_auc_score(y_test,y_pred)
auc_roc
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

Plotting the ROC curve for the classifier

```python
plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

Using Grid Search to find the best paramters for the Logistic Regression model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
LR_model= LogisticRegression()
tuned_parameters = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] ,
              'penalty':['l1','l2']
                   }
from sklearn.model_selection import GridSearchCV
LR= GridSearchCV(LR_model, tuned_parameters,cv=10)
```

Fitting another model with the tuned parameters

```python
LR.fit(X_train,y_train)
```

```python
print(LR.best_params_)
```

```python
y_prob = LR.predict_proba(X_test)[:,1] # This will give you positive class prediction probabilities  
y_pred = np.where(y_prob > 0.5, 1, 0) # This will threshold the probabilities to give class predictions.
LR.score(X_test, y_pred)
```

Computing the confusion matrix for the tuned Logistic Regression model

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

Computing the area under the curve for the tuned model

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

Plotting the ROC curve for the tuned Logistic Regression model

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
